##  Imports

```python
import pandas as pd
import requests
import json
import matplotlib
import numpy as np
%matplotlib inline
```

# Bitcoin market value for last 30 days, vs. USD, deliver a DF and a Plot

```python
base = 'https://api.coingecko.com/api/v3/'
from datetime import datetime
import matplotlib as plt

def get_coin_df(currency):
    url = base + 'coins/'+ str(currency) + '/market_chart?vs_currency=usd&days=30'
    result = requests.get(url)
    j_bc = result.json()
    price_column_label = str(currency) + '_price'
    df_bc = pd.DataFrame(j_bc['prices'], columns=['time', price_column_label])
    df_bc['time'] = df_bc['time'].apply(lambda x: datetime.utcfromtimestamp(int(np.floor(x/1000))).strftime('%Y-%m-%d'))
    return df_bc

df = get_coin_df('bitcoin')

ax = df.plot(x='time', y='bitcoin_price', rot=90)
```

# Litecoin market value for last 30 days, vs. USD, deliver a DF and a Plot

```python
df = get_coin_df('litecoin')

ax = df.plot(x='time', y='litecoin_price', rot=90)
```

## Plot Litecoin and Dogecoin prices in USD in a lineplot for each for the last 30 days

```python
import matplotlib.pyplot as plt

df_litecoin = get_coin_df('litecoin')
df_dogecoin = get_coin_df('dogecoin')
fig, ax = plt.subplots()
ax.plot(df_litecoin['time'], df_litecoin['litecoin_price'])
ax.plot(df_dogecoin['time'], df_dogecoin['dogecoin_price'])
plt.ylabel('Price in USD')
ax.legend(['litecoin', 'dogecoin'])
ticks = plt.xticks(rotation=90)
```

# Top Ten Coins Report

## List the first 10 coingecko currencies in a DataFrame

```python
base = 'https://api.coingecko.com/api/v3/'


url = base + 'coins/markets?vs_currency=usd'
result = requests.get(url)
j_bc = result.json()

top_currencies = pd.DataFrame(data=[],columns=['currency', 'price'])
for currency in j_bc:
    top_currencies = top_currencies.append(
        {'currency':currency['id'], 'price':currency['current_price']},
    ignore_index=True,
    )
    #print(top_currencies)

top_currencies.sort_values(by=['price'], ascending=False).head(10)
```
